Write a Kafka producer that will read “1500000SalesRecords.csv” file line by line and convert each line to JSON and send the data to a Kafka topic. The name of the topic should be “SalesRecords”. 

 

```python
from kafka import KafkaProducer
import json
```

```python
bootstrap_server=['localhost:9092']
topic='SalesRecords'
```

```python
producer=KafkaProducer(bootstrap_servers=bootstrap_server)
```

```python
with open('1500000+Sales+Records.csv', 'r') as f:
   header= f.readline()
   data= f.readlines()
```

```python
header=header.split(',')
header[-1]=header[-1].strip()
```

```python
for lines in data:
    line= lines.split(',')
    line[-1]=line[-1].strip()
    json_data=json.dumps(dict(zip(header, line))).encode('utf-8')
    ack= producer.send(topic, json_data)
    meta=ack.get()
    print(meta.offset)
    
```

```python
# display data on the consumer
bin/kafka-console-consumer.sh 
--bootstrap-server localhost:9092 
--topic SalesRecords
--from-beginning
```
